# **Gráficos de líneas**

El gráfico de líneas se utiliza para representar una relación entre dos datos X e Y en un eje diferente.

Para crear un gráfico se necesitan, como era de esperar, datos.

En este ejemplo vamos a simularos con un vector con numpy, y luego vamos a graficar dos gráficos, una para el seno y otra para el coseno, para ver cómo se muestra:

````python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)

# graficamos en el eje X el vector
# en el eje Y el seno
plt.plot(x, np.sin(x))

# graficamos en el eje X el vector
# en el eje Y el coseno
plt.plot(x, np.cos(x))

# Se muestran ambos en el mismo gráfico
plt.show()
````

Se puede cambiar el formato de las líneas, añadiendo un tercer parámetro a cada plot, como vemos en este ejemplo:

````python
# utilizamos los mismos datos
plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '--')
plt.show()
````

Si se desea, se puede guardar la figura en un archivo, como png y otro formato.

Primero se crea una figura del gráfico, con **.figure()**. Después, se puede guardar esta figura con **.savefig()**.




````python
import numpy as np
x = np.linspace(0, 10, 100)
# guardamos el gráfico en una figura con .figure
fig = plt.figure()
plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '--');

fig.savefig('my_fig.png')
````

**Varios gráficos en uno**

Vamos a ver cómo podemos crear varios gráficos en una misma pantalla, lo que nos permite poder tener una imagen de cada uno de los gráficos y comparar o visualizar diferentes valores.

````python
plt.figure() # create a plot figure
# create the first of two panels and set current axis
plt.subplot(2, 1, 1) # (rows, columns, panel number)
plt.plot(x, np.sin(x))

# create the second panel and set current axis
plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x))

plt.show()
````

También se puede tratar a los gráficos como objetos, almacenando todos en un objeto y llamando a cada uno según el orden que queramos.

En el siguiente código, se obtienen la figura y un array de los 2 ejes del objeto para esa imagen.

Luego se llama con la posición de ese array a cada gráfico.

````python
# Creamos subplots y obtenemos la figura y los ejes
fig, ax = plt.subplots(2)

# Llamamos a cada gráfico con la posición en el array ax
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x));

plt.show()
````

**Ajustando el estilo y colores de los gráficos**

El primer ajuste que se puede hacer a un gráfico es controlar los colores y estilos de las líneas.

 La función **plt.plot()** toma argumentos adicionales que pueden ser utilizados para especificar estos.

Para ajustar el color, puede utilizar la palabra clave *color*, que acepta un argumento que representa prácticamente cualquier color imaginable.

````python
plt.plot(x, np.sin(x - 0), color='blue') # color por el nombre
plt.plot(x, np.sin(x - 1), color='g') # inicial del color
plt.plot(x, np.sin(x - 2), color='0.75') # Grayscale entre 0 y 1
plt.plot(x, np.sin(x - 3), color='#FFDD44') # Hex code
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) # RGB entre 0 y 1
plt.plot(x, np.sin(x - 5), color='chartreuse'); # HTML
````

**Estilos**

De forma similar a los colores, puedes especificar el estilo de las líneas:

````python
plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted');

# For short, you can use the following codes:
plt.plot(x, x + 4, linestyle='-') # solid
plt.plot(x, x + 5, linestyle='--') # dashed
plt.plot(x, x + 6, linestyle='-.') # dashdot
plt.plot(x, x + 7, linestyle=':'); # dotted
````

**Ajustar el límite de los ejes**

Aunque normalmente Matplotlib ajusta bastante bien el límite de los ejes, es bueno poder tener un control para que seamos nosotros quien elijamos esos límites.

La forma de hacer esto es con **plt.xlim()** y **plt.ylim()**:

````python
plt.plot(x, np.sin(x))
# eje X entre -1 y 11
plt.xlim(-1, 11)
# eje y entre -1.5 y 1.5
plt.ylim(-1.5, 1.5)
plt.show()
````

**Usando marcadores**

Además de los colores, los ejes y las líneas, se pueden usar marcadores para diferenciar entre series de un mismo gráfico.

Para crear estos marcadores, se utiliza *market*, dándole el valor que queramos:

````python
plt.plot(x, np.sin(x), marker='o')
plt.plot(x, np.sin(x) + 0.1, marker='.')
plt.plot(x, np.sin(x) + 0.2, marker=',')
plt.plot(x, np.sin(x) + 0.3, marker='v')
plt.plot(x, np.sin(x)+ 0.4, marker='+')

plt.xlim(-1, 5)
# eje y entre -1.5 y 1.5
plt.ylim(-1.5, 1.5)
plt.show()

plt.show()
````

**Etiquetas**

Una de las partes más importantes de los gráficos son el título, los ejes y las leyendas.

Para mostrar el nombre del gráfico se muestra utilizando

```
plt.title('Nombre')
```

Si lo que queremos es mostrar el título de los ejes, se escribe como

```
plt.xlabel('eje x')
plt.ylabel('eje y')
```
Por último, para mostrar la leyenda, se realiza con el método .legend(), introduciendo una lista de valores para cada serie:

```
plt.legend(['Serie 1', 'Serie 2'])
```





**Ejemplo GDP vs Años**

En el siguiente gráfico se muestra un ejemplo de gráfico de líneas, en el que se representa los valores de GDP frente a los años:

````python
# importamos pyplot de matplotlib
from matplotlib import pyplot as plt

# creamos datos
years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gdp = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3]

# creamos un gráfico de líneas
# años en eje x y gdp en eje y
plt.plot(years, gdp, color='green',marker='o',linestyle='solid')
plt.legend(['GDP'])
#añadimos un título
plt.title('Nominal GDP')

# añadimos etiqueta en el eje y
plt.ylabel('Billions of $')
plt.xlabel('Years')
# mostramos el gráfico
plt.show()
````
