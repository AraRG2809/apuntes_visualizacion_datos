#Ejercicio 3

Grafica como evoluciona cada uno de los delitos con el paso del tiempo (desde la columna 'Violent' hasta el final). Realiza los gráficos en una sola pantalla en 3 filas y 3 columnas, rellenándolo de izquierda a derecha.

```python
crime_columns = ['Violent', 'Property', 'Murder', 'Forcible_Rape', 'Robbery',
                 'Aggravated_assault', 'Burglary', 'Larceny_Theft', 'Vehicle_Theft']

# Crear una figura con 3 filas y 3 columnas
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(18, 12))
fig.suptitle('Evolución de cada tipo de delito (1960-2014)', fontsize=16)

# Graficar cada delito
for i, column in enumerate(crime_columns):
    row, col = divmod(i, 3)
    axes[row, col].plot(crime['Year'], crime[column], marker='o')
    axes[row, col].set_title(column)
    axes[row, col].set_xlabel('Year')
    axes[row, col].set_ylabel('Cases')
    axes[row, col].grid(True)

# Ajustar el espacio entre gráficos
plt.tight_layout(rect=[0, 0, 1, 1])

# Mostrar los gráficos
plt.show()
```

c) A partir del 2000 los crímenes tienen una tendencia decreciente

```python
solucion_3="c"
```

#Ejercicio 4

Crea la función solucion_4 que devuelva el gráfico donde se representa:

* los crimenes de asesinato 'Muder' con una línea de puntos roja y con marcador de triángulos.

* Los crímenes de robos 'Robbery' con una linea verde discontinua y marcador de puntos.

* Los crímenes de robo de vehículo 'Vehicle_Theft' con una línea azul continua y marcador de cruces.

El gráfico debe contener como leyenda 'asesinatos', 'robos en domicilios' y 'coches robados'; etiqueta del eje x 'Año' y del eje y 'Delitos'.

```python

def solucion_4():

    # Graficar los crímenes de asesinato (Murder)
    plt.plot(crime['Year'], crime['Murder'], 'r--^', label='asesinatos')

    # Graficar los crímenes de robos (Robbery)
    plt.plot(crime['Year'], crime['Robbery'], 'g-.o', label='robos en domicilios')

    # Graficar los crímenes de robo de vehículo (Vehicle_Theft)
    plt.plot(crime['Year'], crime['Vehicle_Theft'], 'b-x', label='coches robados')

    # Añadir leyenda
    plt.legend()

    # Etiquetas de los ejes
    plt.xlabel('Año')
    plt.ylabel('Delitos')

    # Título del gráfico (opcional)
    plt.title('Evolución de crímenes en EE.UU. (1960-2014)', fontsize=16)

    # Mostrar el gráfico
    plt.grid(True)
    plt.show()

# Llamar a la función para generar el gráfico
solucion_4()
```
