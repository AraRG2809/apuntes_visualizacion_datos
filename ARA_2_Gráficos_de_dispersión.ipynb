## **Gráficos de dispersión**

Los gráficos de dispersión se utilizan para observar las relaciones entre las variables y utiliza puntos para representar la relación entre ellas.

Otro tipo de gráfico comúnmente utilizado es el gráfico de dispersión simple, un gráfico similar al gráfico de líneas.

En lugar de que los puntos estén unidos por segmentos de línea, aquí los puntos se representan individualmente con un punto, un círculo u otra forma.

Comenzaremos importando las librerías:

```python
import matplotlib.pyplot as plt
import numpy as np
```

Vamos a crear un gráfico de dispersión ejemplo. Se crea de forma similar a los gráficos de líneas, pero hay que introducir un parámetro nuevo que se corresponda con el marcador, ya que, por defecto, incluye el de líneas:

```python
x = np.linspace(0, 10, 30)
y = np.sin(x)

plt.plot(x, y, 'o')
plt.show()
```

Otra forma de crear el gráfico de dispersión es mediante *plt.scatter()* en lugar de usar la función plot:

```python
plt.scatter(x, y, marker='o')
plt.show()
```

**Estilos de gráficos de dispersión con plt.scatter**

La principal diferencia de *plt.scatter* con respecto a *plt.plot* es que puede utilizarse para crear gráficos de dispersión en los que las propiedades de cada punto individual (tamaño, color de la cara, color del borde, etc.) pueden controlarse individualmente o asignarse a los datos.

Vamos a demostrarlo creando un gráfico de dispersión aleatorio con puntos de muchos colores y tamaños.

Para ver mejor los resultados superpuestos, también usaremos la palabra clave alfa para ajustar el nivel de transparencia:

```python
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)
plt.scatter(x, y, c=colors, s=sizes, alpha=0.3,
            cmap='viridis')
plt.colorbar()
plt.show()
```

En el gráfico anterior, el color y el tamaño de los puntos se asigna automáticamente según el valor en la escala.

De este modo, el color y el tamaño de los puntos pueden utilizarse para transmitir información en la visualización, con el fin de ilustrar datos multidimensionales.

**Ejemplo Dataset Iris**

Vamos a representar datos de 3 tipos de flores con diferten tamaños de pétalos y sépalos:

```python
# importamos datos de sklearn.datasets
from sklearn.datasets import load_iris
# cargamos los datos
iris = load_iris()
# los representamos con la trans
features = iris.data.T

iris
```

Una vez hemos obtenido los datos que queremos representar, vamos a crear el gráfico de dispersión.

Para crear el gráfico de dispersión vamos a usar *plt.scatter()*.

Vamos a representar las dos primeras columnas de los datos.

Le incorporaremos los nombres a los ejes X e Y.

También al título.


```python
plt.scatter(features[0], features[1], alpha=0.2,
            s=100*features[3], c=iris.target, cmap='viridis')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.title('Iris')
plt.show()
```

Podemos ver que este gráfico de dispersión nos ha dado la posibilidad de explorar simultáneamente cuatro dimensiones diferentes de los datos:
* la ubicación (x, y) de cada punto corresponde a la *longitud* y la *anchura* del *sépalo*
* el tamaño del punto está relacionado con la *anchura del pétalo*
* el color está relacionado con la *especie particular de la flor*.

Los gráficos de dispersión multicolores y de múltiples características como éste pueden ser útiles tanto para la exploración como para la presentación de los datos.

De esta forma, podemos observar diferentes grupos de flores, en función de la anchura y longitud del sépalo, obteniendo de una forma visual los resultados.
