## **Gráficos de barras**

Los gráficos de barras son una de las mejores técnicas de visualización que están disponibles en la biblioteca matplotlib para que los usuarios las utilicen.

Estos gráficos representan la categoría de los datos con barras rectangulares con longitudes y alturas proporcionales a los valores que representan.

Se puede crear utilizando el método **bar()**.

Estos gráficos se utilizan a menudo para comparar elementos de datos presentes en los conjuntos de datos.

También es extremadamente útil para entender el tipo de datos y averiguar si los datos están balanceados o no (si tienen el mismo número de datos por cada tipo).

* Para crear gráficos verticales, se utiliza **plt.bar()**
* Para crear gráficos horizontales, se hace mediante plt.barh()

Se pueden crear gráficos utilizando matplotlib o directamente desde pandas.

Mostramos a continuación un ejemplo utilizando matplotlib.

## **Gráficos de barras con matplotlib**

```python
# importamos librerías
import matplotlib.pyplot as plt
import numpy as np

# creamos datos con numpy
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

# utilizamos subplot para graficar
plt.subplot(2,1,1) # nrows, ncols, index
plt.bar(x,y)

# utilizamos subplot para graficar
plt.subplot(2,1,2) # nrows, ncols, index
plt.barh(x,y)

# mostramos gráfico
plt.show()
```

## **Gráficos de barras desde Pandas**

En este caso se utiliza el método plot de pandas.

```python
# importamos las librerías
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# generamos la figura y los ejes con 2 gráficos
fig, axes = plt.subplots(2, 1)

# generamos datos aleatorios
data = pd.Series(np.random.rand(16), index=list('abcdefghijklmnop'))

# gráficos de barras vertical utilizando plot de pandas
data.plot.bar(ax=axes[0], color='b', alpha=0.7)

# gráfico de barras horizontal
data.plot.barh(ax=axes[1], color='b', alpha=0.7)
plt.show()
```

## **Gráficos de barras agrupados**

Con un DataFrame, los gráficos de barras agrupan los valores de cada fila en un grupo en barras, uno al lado del otro, para cada valor.

```python
# Creamos DataFrame ejemplo
df = pd.DataFrame(np.random.rand(6, 4),
                  index=['one', 'two', 'three', 'four', 'five', 'six'],
                  columns=pd.Index(['A', 'B', 'C', 'D'], name='Genus'))
```

```python
# Mostramos dataframe
df.head()
```

```python
# Creamos gráfico con cada grupo
df.plot.bar()
plt.show()
```

## **Gráficos apilados**

Es posible crear gráficos de barras apilados, de forma que en una misma barra podemos representar diferentes series.

Para crear esto se puede utilizar Pandas, creando gráficos de barras apiladas a partir de un DataFrame pasando el argument *stacked=True*, lo que permite que los valores de cada fila se apilen juntos.


```python
df.plot.barh(stacked=True, alpha=0.5)
plt.show()
```
