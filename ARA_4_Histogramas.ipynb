## **Histogramas**

Un histograma es un gráfico que muestra las distribuciones de frecuencia.

Es un gráfico que muestra el número de observaciones dentro de cada intervalo dado.

Estas visualizaciones de histogramas se utilizan mejor para el trazado de distribuciones de frecuencia.

Los histogramas se utilizan para generar imágenes y gráficos precisos de una lista o una matriz en Python. Esta aplicación es especialmente para proyectos de Ciencia de Datos.

Las aplicaciones de los histogramas incluyen el desarrollo de visualizaciones para problemas relacionados con el análisis de series de tiempo y otra previsión de negocios.

Si tienes un patrón de frecuencias largo y estable o un array de elementos de datos, aprovecha esta oportunidad para beneficiarte de la técnica de visualización de histogramas disponible en la librería matplotlib.

Un histograma es un tipo de gráfico de barras que ofrece una visualización discreta de la frecuencia de los valores.

Los puntos de datos se dividen en intervalos discretos, espaciados uniformemente, y se traza el número de puntos de datos en cada intervalo.

Se puede crear un histograma con el método **plot.hist**:

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.normal(170, 10, 250)

plt.hist(x)
plt.show()
```

## **Personalizar histogramas**

La función hist() tiene muchas opciones para ajustar tanto el cálculo como la visualización.

Aquí podéis ver un ejemplo de un histograma más personalizado:

```python
data = np.random.randn(1000)
plt.hist(data, bins=50, alpha=0.5,
                    histtype='stepfilled', color='steelblue',
                    edgecolor='none')
plt.show()
```

# **Bonus: Gráfico de densidad**

Un tipo de gráfico relacionado es el gráfico de densidad, que se forma calculando una estimación de una **distribución de probabilidad continua** que podría haber generado los datos observados.

El procedimiento habitual es aproximar esta distribución como una mezcla de "núcleos", es decir, distribuciones más simples como la distribución normal.

Por lo tanto, los gráficos de densidad también se conocen como gráficos de estimación de densidad del núcleo (KDE).

Con Seaborn, puedes utilizar **plot.kde** se obtiene un gráfico de densidad utilizando la estimación convencional de la mezcla de normales

```python
import seaborn as sns

#define data
x = np.random.normal(170, 10, 250)

#create density plot of data
sns.kdeplot(x)
plt.show()
```
