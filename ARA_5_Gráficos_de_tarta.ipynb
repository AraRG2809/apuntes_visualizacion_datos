## **Gráficos de tarta**

El **gráfico circular** o de tarta o queso es una forma brillante de visualizar los elementos de datos en su conjunto de datos.

Esta técnica de visualización se despliega en los proyectos de Ciencia de Datos para averiguar sus cálculos y cómputos en **porcentajes**. Esta técnica es especialmente útil para determinar un enfoque porcentual en sus conjuntos de datos.

El gráfico circular también puede utilizarse para **comparar las numerosas distribuciones de los datos** y elementos de datos. Sin embargo, es importante tener en cuenta que las rebanadas producidas en el gráfico circular tienen espacios variables. De lo contrario, la visualización podría resultar compleja porque todos los porcentajes estarían más cerca unos de otros de lo esperado.

Vemos un ejemplo:

```python
# importar librerías
from matplotlib import pyplot as plt
import numpy as np


# crear dataset
cars = ['AUDI', 'BMW', 'FORD',
		'TESLA', 'JAGUAR', 'MERCEDES']

data = [23, 17, 35, 29, 12, 41]

# crear gráfico
fig = plt.figure(figsize =(10, 7))
plt.pie(data, labels = cars)

# mostrar gráfico
plt.show()
```

En el gráfico anterior, se puede observar una comparación entre las diferentes marcas de coches.

Suponiendo que represente el % del mercado que acaparan, podría decirse que Mercedes es la que más mercado tiene, seguido de Ford, Tesla y Audi.

**Pros:** Este gráfico nos permite comparar de una forma visual diferentes conjuntos de datos, y así ver a simple vista cuáles son más representativos.

**Contras:** No son muy matemáticos y se recomienda no usarlos, o, al menos, acompañarlos con su %, ya que de forma visual es complicado saber realmente la diferencia real(Ej: Tesla vs Ford).

## **Customizar un gráfico circular**
Podemos incluir los porcentajes u otros datos y hacerlo más personalizado con más parámetros:

```python
import matplotlib.pyplot as plt

cars = ['AUDI', 'BMW', 'FORD',
		'TESLA', 'JAGUAR', 'MERCEDES']

data = [23, 17, 35, 29, 12, 41]

plt.pie(data, labels=cars, autopct='%1.1f%%', explode=[0,0,0,0,0,0.1], shadow=True, startangle=90)
plt.title('Cuota de mercado vehículos de lujo en USA')
plt.axis('equal')
plt.show()
```
