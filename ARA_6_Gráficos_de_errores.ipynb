## **Visualización de errores**

Para cualquier medición, contabilizar los errores de forma precisa es casi tan importante como los propios análisis.

Por ejemplo, imaginemos que utilizamos algunas observaciones astrofísicas para estimar la Constante de Hubble, la medida local de la velocidad de expansión del universo.

Sé que la literatura actual sugiere un valor de alrededor de 71 (km/s)/Mpc, y nosotros medimos un valor de 74 (km/s)/Mpc con mi método.

¿Son coherentes los valores?

La única respuesta correcta, dada esta información, es ésta: no hay forma de saberlo.

Supongamos que se aumenta esta información con las incertidumbres reportadas: la literatura actual sugiere un valor de alrededor de 71 ± 2,5 (km/s)/Mpc, y mi método ha medido un valor de 74 ± 5 (km/s)/Mpc.

Ahora bien, ¿son los valores coherentes?

Esa es una pregunta que se puede responder cuantitativamente.

En la visualización de datos y resultados, mostrar estos errores de forma efectiva puede hacer que un gráfico transmita una información mucho más completa.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('seaborn-whitegrid')

x = np.linspace(0, 10, 50)
dy=0.8
y = np.sin(x) + dy * np.random.randn(50)
plt.errorbar(x, y, yerr=dy, fmt='.k');
```

Dada la gráfica anterior, se pueden visualizar los errores en cada punto, obteniendo mayor información sobre las estimaciones y permitiendo realizar un análisis más profundo.

En este caso, cada punto cuenta con un error similar, y también lo hace tanto por arriba como por abajo, lo cual no suele ser lo habitual y ocurre porque no hemos utilizado datos reales.
