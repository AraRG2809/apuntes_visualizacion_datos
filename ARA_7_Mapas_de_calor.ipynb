## **Mapa de calor**

Los mapas de calor son una de las técnicas de visualización más **útiles para el cálculo de problemas complejos relacionados con el ajuste de hiperparámetros** en la construcción y el análisis de algoritmos de machine learning como los árboles de decisión, los bosques aleatorios, los KNN, y otros algoritmos que requieren el uso de averiguar los valores precisos de los hiperparámetros.

Estas visualizaciones de mapas de calor a menudo utilizan tanto la **biblioteca matplotlib** como la **biblioteca seaborn** para la interpretación de los valores precisos para determinar los mejores valores para la tarea particular.

Los mapas de calor producen **interpretaciones multicolores** que ayudan a obtener una comprensión más detallada de los elementos de los datos.

Se disponen los **datos de una forma matricial**, siendo el eje X una variable y el eje Y otra variable, y en la intersección de dos valores se dispone un color según el valor en una escala.

El mapa de calor en sí es un gráfico *imshow* con las etiquetas establecidas en las categorías que tenemos.

Podemos ver un simple ejemplo a continuación, utilizando **imshow**:

```python
import matplotlib.pyplot as plt
import numpy as np

a = np.random.random((16, 16))
plt.imshow(a, cmap='hot', interpolation='nearest')
plt.show()
```

Este tipo de gráficos se utilizan para ver la correlación, o la importancia entre variables, teniendo un conjunto de variables en un eje y otro conjunto en otro eje.

Un ejemplo en el que se pueden utilizar los mapas de calor es para representar la matriz de confusión en un algoritmo de clasificación.

En el siguiente gráfico se puede ver la probabilidad de acierto que tiene un algoritmo cuando clasifica una imagen como 0 o 1 y realmente es ese resultado.

En el eje Y los valores que etiqueta el algoritmo, y en el eje X los valores reales del algoritmo.
