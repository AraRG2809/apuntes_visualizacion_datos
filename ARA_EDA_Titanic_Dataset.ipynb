# Titanic dataset

Datos de los pasajeros del titanic.

Importamos librerías

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Cargamos datos

```python
# Cargar el conjunto de datos del Titanic
url = "https://web.stanford.edu/class/archive/cs/cs109/cs109.1166/stuff/titanic.csv"
titanic_data = pd.read_csv(url)
```

Exploramos primeras filas

```python
# Explorar los primeros registros del conjunto de datos
titanic_data.head()
```

Resumen estadístico de los datos

```python
# Resumen estadístico del conjunto de datos
titanic_data.describe()
```

Distribución de la superviviencia

```python
# Gráfico de barras para visualizar la distribución de supervivencia
sns.countplot(x='Survived', data=titanic_data)
plt.title('Distribución de Supervivencia')
plt.show()
```

Distribución de clases

```python
# Gráfico de barras para la distribución de clases
sns.countplot(x='Pclass', data=titanic_data)
plt.title('Distribución de Clases')
plt.show()
```

Barras apiladas para distribución de clases por supervivencia

```python
# Gráfico de barras apiladas para la distribución de clases por supervivencia
sns.countplot(x='Pclass', hue='Survived', data=titanic_data)
plt.title('Distribución de Clases por Supervivencia')
plt.show()
```

Ver la relación entre la edad, la tarifa y los supervivientes

```python
# Gráfico de dispersión para la relación entre la edad y la tarifa
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Fare', data=titanic_data, hue='Survived', palette='viridis')
plt.title('Relación entre Edad y Tarifa con Supervivencia')
plt.show()
```

Distribución de supervivencia por sexo

```python
# Gráfico de barras apiladas para la distribución de clases por supervivencia
sns.countplot(x='Sex', hue='Survived', data=titanic_data)
plt.title('Distribución de Supervivencia por Sexo')
plt.show()
```

Gráfico de densidad para la edad

```python
# Gráfico de densidad para la distribución de la edad
plt.figure(figsize=(10, 6))
sns.kdeplot(titanic_data['Age'], fill=True)
plt.title('Distribución de Edades')
plt.show()
```
