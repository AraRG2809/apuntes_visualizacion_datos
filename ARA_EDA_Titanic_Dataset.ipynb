# Titanic dataset

Datos de los pasajeros del titanic.

Importamos librer√≠as

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Cargamos datos

```python
# Cargar el conjunto de datos del Titanic
url = "https://web.stanford.edu/class/archive/cs/cs109/cs109.1166/stuff/titanic.csv"
titanic_data = pd.read_csv(url)
```

Exploramos primeras filas

```python
# Explorar los primeros registros del conjunto de datos
titanic_data.head()
```

Resumen estad√≠stico de los datos

```python
# Resumen estad√≠stico del conjunto de datos
titanic_data.describe()
```

Distribuci√≥n de la superviviencia

```python
# Gr√°fico de barras para visualizar la distribuci√≥n de supervivencia
sns.countplot(x='Survived', data=titanic_data)
plt.title('Distribuci√≥n de Supervivencia')
plt.show()
```

Distribuci√≥n de clases

```python
# Gr√°fico de barras para la distribuci√≥n de clases
sns.countplot(x='Pclass', data=titanic_data)
plt.title('Distribuci√≥n de Clases')
plt.show()
```

Barras apiladas para distribuci√≥n de clases por supervivencia

```python
# Gr√°fico de barras apiladas para la distribuci√≥n de clases por supervivencia
sns.countplot(x='Pclass', hue='Survived', data=titanic_data)
plt.title('Distribuci√≥n de Clases por Supervivencia')
plt.show()
```

Ver la relaci√≥n entre la edad, la tarifa y los supervivientes

```python
# Gr√°fico de dispersi√≥n para la relaci√≥n entre la edad y la tarifa
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Fare', data=titanic_data, hue='Survived', palette='viridis')
plt.title('Relaci√≥n entre Edad y Tarifa con Supervivencia')
plt.show()
```

Distribuci√≥n de supervivencia por sexo

```python
# Gr√°fico de barras apiladas para la distribuci√≥n de clases por supervivencia
sns.countplot(x='Sex', hue='Survived', data=titanic_data)
plt.title('Distribuci√≥n de Supervivencia por Sexo')
plt.show()
```

Gr√°fico de densidad para la edad

```python
# Gr√°fico de densidad para la distribuci√≥n de la edad
plt.figure(figsize=(10, 6))
sns.kdeplot(titanic_data['Age'], fill=True)
plt.title('Distribuci√≥n de Edades')
plt.show()
```
